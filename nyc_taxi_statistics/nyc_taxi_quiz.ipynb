```python
import pandas as pd
```

```python
taxi10000 = pd.read_csv('sample10000.csv', parse_dates=[1,2])
taxi10000.head()
```

```python
numbCachePayer = len(taxi10000.loc[taxi10000['payment_type'] == 2])
```

```python
cachePayerProportion = float(len(taxi10000.loc[taxi10000['payment_type'] == 2])) / len(taxi10000)
print cachePayerProportion
```

```python
from statsmodels.stats.proportion import proportion_confint
proportion_confint(numbCachePayer, len(taxi10000), alpha=0.01)
```

```python
avgTripDistance = taxi10000['trip_distance'].mean()
print avgTripDistance
```

```python
import numpy as np
s = taxi10000['trip_distance'].std(ddof=1) / np.sqrt(len(taxi10000['trip_distance']))
s
```

```python
from statsmodels.stats.weightstats import _tconfint_generic
_tconfint_generic(taxi10000['trip_distance'].mean(), 
                  s, 
                  len(taxi10000['trip_distance']) - 1, 
                  0.05, 'two-sided')
```
